```python
old_mean = np.mean(ins_data['Old Scheme (in thousands)'])
new_mean = np.mean(ins_data['New Scheme (in thousands)'])

```

```python
from scipy.stats import ttest_1samp,ttest_ind
t_stat,p = ttest_ind(ins_data['Old Scheme (in thousands)'] ,ins_data['New Scheme (in thousands)'])
```

```python
print(t_stat,p)
```

```python
if(p < 0.05):
    print("2. Null Hypothesis : There is no significant difference between data sets")
else:
    print ("2. Alternate Hypothesis : There is a significant difference")
```

```python
if(p < 0.05):
    print("3. As it is Null Hypothesis, we can conclude that new scheme doen't have any significantly raised outputs.")
else:
    print("3. As we reject the Null Hypothesis, we can conclude that new scheme has significantly raised outputs.")
```

```python
# Compute the difference in Means between 2 sample means and divide by pooled Standard Deviation 
# number of Observations/tuples
# Set the alpha value to 0.05 and alternative values 'two-sided' , 'larger' , 'smaller'

newScheme = ins_data['New Scheme (in thousands)']
oldScheme =ins_data['Old Scheme (in thousands)']
effect_size=(np.mean(newScheme) - np.mean(oldScheme)) / np.sqrt(((30-1)*np.var(newScheme)+(30-1)*np.var(oldScheme)) / 30+30-2)
print(effect_size)
```

```python
# calculate the t test
from statsmodels.stats.power import ttest_power
alpha = 0.05
powerOfTest = ttest_power(effect_size, nobs=30, alpha=alpha, alternative='two-sided')
print("4. c)Power of Test Value: {0}".format(powerOfTest))
```
