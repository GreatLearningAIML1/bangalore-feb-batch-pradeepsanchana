```python
import numpy as np
import pandas as pd
```

```python
ins_data = pd.read_csv("data.csv")
```

```python
old_mean = np.mean(ins_data['Old Scheme (in thousands)'])
old_mean
```

```python
new_mean = np.mean(ins_data['New Scheme (in thousands)'])
new_mean
```

```python
print("Old mean:",  round(old_mean,2))
print("New mean:" , round(new_mean,2))
print("Old std:",  round(np.std(ins_data['Old Scheme (in thousands)']),2))
print("New std:" , round(np.std(ins_data['New Scheme (in thousands)']),2))

```

```python
from scipy.stats import ttest_1samp,ttest_ind
t_stat,p = ttest_ind(ins_data['Old Scheme (in thousands)'] ,ins_data['New Scheme (in thousands)'])
```

```python
print(t_stat,p)
```

```python
if(p < 0.05):
    print(" Null Hypothesis : There is no significant difference between data sets")
else:
    print ("Alternate Hypothesis : There is a significant difference")
```

```python
if(p < 0.05):
    print("As it is Null Hypothesis, we can conclude that new scheme doen't have any significantly raised outputs.")
else:
    print("As we reject the Null Hypothesis, we can conclude that new scheme has significantly raised outputs.")
```

```python
print ('As we are using Null hypothesis test, the significance of the alpha is the probability of Type 1 error. which is 0.05')
```
